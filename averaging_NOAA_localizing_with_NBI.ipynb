```python
import os

weather_stns_list = []

directory = r'C:\Users\tobby\Documents\Descriptive_Predictive_Bridge_Management\El_Paso_Weather_Stations'
for filename in os.listdir(directory):
    current_year = filename.split('.')[0].split('_')[2]
#     print(filename.split('.')[0].split('_')[2])
    with open(os.path.join(directory, filename)) as fp:
        line = fp.readline()
        count = 1
        while line:
            weather_stns_list.append((current_year, line.split(',')[0]))
            print('line {}: {}'.format(count, line.strip()))
            line = fp.readline()
            count += 1
```

```python
weather_stns_list
```

```python
list_of_avg_data = []

counter = 0
for weather_stn in weather_stns_list:
    counter += 1
    print(counter)
    
    count = 0
    prcp_total = 0
    snow_total = 0
    snwd_total = 0
    
#     try:
    
    for idx, row in df_trimmed.iterrows():

        if  weather_stn[0] in str(row['DATE']) and row['STATION'] == weather_stn[1]:
            count += 1
            prcp_total += row['PRCP']
#             snow_total += row['SNOW']
#             snwd_total += row['SNWD']

    if count != 0:
        prcp_avg = prcp_total/count
        snow_avg = snow_total/count
        snwd_avg = snwd_total/count

        list_of_avg_data.append((weather_stn[1], weather_stn[0], prcp_avg, snow_avg, snwd_avg))
#     except:
#         continue
```
