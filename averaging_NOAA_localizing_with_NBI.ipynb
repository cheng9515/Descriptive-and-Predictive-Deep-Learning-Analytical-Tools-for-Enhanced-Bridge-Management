```python
import os

list_of_dfs = []
directory = r'C:\Users\tobby\Documents\Descriptive_Predictive_Bridge_Management\El_Paso_NOAA_data'
for filename in os.listdir(directory):
    temp_df = pd.read_csv(os.path.join(directory, filename))
    list_of_dfs.append(temp_df)
```

```python
df_concat = pd.concat(list_of_dfs, ignore_index=True)
```

```python
df_concat = df_concat[['STATION', 'DATE', 'LATITUDE', 'LONGITUDE', 'PRCP', 'SNOW', 'SNWD']].copy()
```

```python
list_of_id_year_tuples = []
for idx, row in df_concat.iterrows():
    list_of_id_year_tuples.append((row['STATION'], row['DATE'].split('-')[0], row['LATITUDE'], row['LONGITUDE']))
```

```python
set_of_id_year_tuples = set()
for id_year_tuple in list_of_id_year_tuples:
    set_of_id_year_tuples.add(id_year_tuple)
```

```python
len(list(set_of_id_year_tuples))
```

```python
list_of_avg_data = []
counter = 0

for id_year_tuple in list(set_of_id_year_tuples):
    counter += 1
    print(counter)
    
#     if counter > 5:
#         break
    
    count = 0
    prcp_total = 0
    snow_total = 0
    snwd_total = 0
    
    for idx, row in df_concat.iterrows():
        
        if id_year_tuple[1] in str(row['DATE']) and row['STATION'] == id_year_tuple[0]:
            count += 1
            prcp_total += row['PRCP']
            snow_total += row['SNOW']
            snwd_total += row['SNWD']
        
    if count != 0:
        prcp_avg = prcp_total/count
        snow_avg = snow_total/count
        snwd_avg = snwd_total/count
        
        list_of_avg_data.append((id_year_tuple[0], id_year_tuple[1], prcp_avg, snow_avg, snwd_avg))
        
        
# list_of_avg_data = []

# counter = 0
# for weather_stn in weather_stns_list:
#     counter += 1
#     print(counter)
    
#     count = 0
#     prcp_total = 0
#     snow_total = 0
#     snwd_total = 0
    
# #     try:
    
#     for idx, row in df_trimmed.iterrows():

#         if  weather_stn[0] in str(row['DATE']) and row['STATION'] == weather_stn[1]:
#             count += 1
#             prcp_total += row['PRCP']
# #             snow_total += row['SNOW']
# #             snwd_total += row['SNWD']

#     if count != 0:
#         prcp_avg = prcp_total/count
#         snow_avg = snow_total/count
#         snwd_avg = snwd_total/count

#         list_of_avg_data.append((weather_stn[1], weather_stn[0], prcp_avg, snow_avg, snwd_avg))
# #     except:
# #         continue
```
